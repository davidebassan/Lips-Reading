## Implementation of VGG16 on Keras

VGG16 is a convolution neural net (CNN) architecture which was used to win ILSVR(Imagenet) competition in 2014


```python
# Importing Libraries
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.utils import load_img, img_to_array
```

```python
img_raw = load_img('cat.jpeg', target_size=(224,224))
img = img_to_array(img_raw)
print(img.shape)
img = img.reshape((1, img.shape[0], img.shape[1], img.shape[2]))
print(img.shape)
```

```python
img_raw
```

```python
model = VGG16()
```

```python
model.summary()
```

```python
img = preprocess_input(img)
y_hat = model.predict(img)
```

```python
decoded_prediction = decode_predictions(y_hat)
print(decoded_prediction[0][0])
```
